```python
from google_play_scraper import app, reviews
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
app_id = 'com.zte.iptvclient.android.idmnc'

# Mengambil ulasan
reviews_data, _ = reviews(
    app_id,
    lang='id',
    country='id',
    count=50000
)

# Konversi ke DataFrame
reviews_data = pd.DataFrame(reviews_data)
reviews_data
```

```python
# save to csv
reviews_data.to_csv('visionplus_reviews.csv', index=False)
```
